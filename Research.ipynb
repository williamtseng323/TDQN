```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("train_data_full.csv")
```

```python
df.iloc[:int(len(df)*0.8),:].to_csv("train_data_80P.csv")
df.iloc[int(len(df)*0.8):,:].to_csv("test_data_20P.csv")
```

```python
fig, ax1 = plt.subplots(figsize=(30, 6))

ax1.bar(df.index, df["macd_signal"], label='Resist')
ax1.set_xlabel('t')
ax1.set_ylabel('Resist', color='b')
ax1.tick_params('y', colors='b')


# Create the second axis sharing the x-axis with the first axis
ax2 = ax1.twinx()

# Plot the second data on the second axis
ax2.plot(df.index, df["close"], 'r-', label='Price')
ax2.set_ylabel('Close', color='r')
ax2.tick_params('y', colors='r')

# Set the legends
# lines = [ax1.get_lines()[0], ax2.get_lines()[0]]
# labels = [line.get_label() for line in lines]
# ax1.legend(lines, labels)
```
